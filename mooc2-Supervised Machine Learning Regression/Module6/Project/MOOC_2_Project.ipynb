# Perform one-hot encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Select encoded columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
print("Columns that will be applied One-Hot Encode:", categorical_cols)

# Initialize encoder
encoder = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')

# Fit và transform data
encoded_array = encoder.fit_transform(df[categorical_cols])

# Get new column names after encoding
encoded_cols = encoder.get_feature_names_out(categorical_cols)

# Create new DataFrame from encoding results
df_encoded_part = pd.DataFrame(encoded_array, columns=encoded_cols, index=df.index)

# Combine encoded data and the rest
df_final = pd.concat([df.drop(columns=categorical_cols), df_encoded_part], axis=1)

print("\n DataFrame apter applying One-Hot Encoding:")
print(df_final.head())
```

# Create Variables and Target features

```python
data = df_final.copy()
y_col = data['price']
X = data.drop(columns='price',axis=0)
Y = data['price']
print('X shape: ',X.shape)
print('Y shape: ',Y.shape)
```

# Create Pipeline and define some superparameters


```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

estimator_lr = Pipeline([("polynomial_features",PolynomialFeatures()),
                         ("scaler",StandardScaler()),
                         ("linear_regression",LinearRegression())])
estimator_las = Pipeline([("polynomial_features",PolynomialFeatures()),
                         ("scaler",StandardScaler()),
                         ("las_regression",Lasso())])
estimator_r = Pipeline([("polynomial_features",PolynomialFeatures()),
                         ("scaler",StandardScaler()),
                         ("ridge_regression",Ridge())])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
```

```python
kf = KFold(shuffle=True, random_state=42, n_splits=3)
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=40)
```

```python
params_r = {'polynomial_features__degree': [1,2,3],
            'ridge_regression__alpha': np.geomspace(0.1,50,20)}
params_las = {'polynomial_features__degree': [1,2,3],
            'las_regression__alpha': np.geomspace(0.1,50,20)}
params_lr = {'polynomial_features__degree': [1,2,3]}
```

```python
from sklearn.model_selection import GridSearchCV
grid_r = GridSearchCV(estimator_r, params_r, cv=kf)
grid_r.fit(X_train, Y_train)

grid_lr = GridSearchCV(estimator_lr, params_lr, cv=kf)
grid_lr.fit(X_train, Y_train)

grid_las = GridSearchCV(estimator_las, params_las, cv=kf)
grid_las.fit(X_train, Y_train)
```

# Assess models

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

def rmse(Y_pred, Y_test):
    return np.sqrt(mean_squared_error(Y_test, Y_pred))

# Predict
Y_pred_lr = grid_lr.predict(X_test)
Y_pred_las = grid_las.predict(X_test)
Y_pred_r = grid_r.predict(X_test)

# Calculate R2 score and RMSE
r2_lr = r2_score(Y_test, Y_pred_lr)
r2_las = r2_score(Y_test, Y_pred_las)
r2_r = r2_score(Y_test, Y_pred_r)

rmse_lr = rmse(Y_pred_lr, Y_test)
rmse_las = rmse(Y_pred_las, Y_test)
rmse_r = rmse(Y_pred_r, Y_test)

# Print results
print("-=Linear Regression=-")
print(f"R2 score: {r2_lr}")
print(f"RMSE: {rmse_lr}\n")

print("-=Lasso Regression=-")
print(f"R2 score: {r2_las}")
print(f"RMSE: {rmse_las}\n")

print("-=Ridge Regression=-")
print(f"R2 score: {r2_r}")
print(f"RMSE: {rmse_r}")
```

```python
import pandas as pd
import numpy as np

# Print degree for each model
print("Best degree (Linear):", grid_lr.best_params_['polynomial_features__degree'])
print("Best degree (Lasso):", grid_las.best_params_['polynomial_features__degree'])
print("Best degree (Ridge):", grid_r.best_params_['polynomial_features__degree'])

# Get feature names from best models
feature_names_lr = grid_lr.best_estimator_.named_steps['polynomial_features'].get_feature_names_out(X.columns)
feature_names_las = grid_las.best_estimator_.named_steps['polynomial_features'].get_feature_names_out(X.columns)
feature_names_r = grid_r.best_estimator_.named_steps['polynomial_features'].get_feature_names_out(X.columns)

# Get coefficents of each model
coef_lr = grid_lr.best_estimator_.named_steps['linear_regression'].coef_
coef_las = grid_las.best_estimator_.named_steps['las_regression'].coef_
coef_r = grid_r.best_estimator_.named_steps['ridge_regression'].coef_

# Create 3 Dataframe for 3 models
df_coef_lr = pd.DataFrame({
    'Feature': feature_names_lr,
    'Coefficient': coef_lr
})

df_coef_las = pd.DataFrame({
    'Feature': feature_names_las,
    'Coefficient': coef_las
})

df_coef_r = pd.DataFrame({
    'Feature': feature_names_r,
    'Coefficient': coef_r
})

# Display basic information
print("\n=== Linear Regression Coefficients ===")
print(df_coef_lr.head())
print(f"Number of coefficients: {len(df_coef_lr)}")

print("\n=== Lasso Regression Coefficients ===")
print(df_coef_las.head())
print(f"Number of coefficients: {len(df_coef_las)}")

print("\n=== Ridge Regression Coefficients ===")
print(df_coef_r.head())
print(f"Number of coefficients: {len(df_coef_r)}")
```
